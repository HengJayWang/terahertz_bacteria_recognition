```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ecoli_file_name = "acesolution_E.Coli_2017_1102/E_Coli_10_5_order_10_ul_x5_feature_vector_data.csv"
yeast_file_name = "acesolution_Yeast_2017_1109/Yeast_10_5_order_10ul_x5_feature_vector_data.csv"
```

```python
ecoli_csv_file = pd.read_csv(ecoli_file_name, low_memory=False)
yeast_csv_file = pd.read_csv(yeast_file_name, low_memory=False)
```

```python
ecoli_feature = ecoli_csv_file.values[:, 1:].transpose()
ecoli_spectral_feature = ecoli_feature[:, 4096:4422]
yeast_feature = yeast_csv_file.values[:, 1:].transpose()
yeast_spectral_feature = yeast_feature[:, 4096:4422]
```

```python
ecoli_feature.shape
yeast_feature.shape
```

```python
ecoli_spectral_feature.shape
yeast_spectral_feature.shape
```

```python
plt.figure(figsize=(50,25))
for i in range(1, 300):
    plt.plot(ecoli_spectral_feature[i, :])
plt.xlabel('N Samples', fontsize=30)
plt.ylabel('Intensity', fontsize=30)
plt.title("E.Coli Spectral Feature", fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(50,25))
for i in range(1, 300):
    plt.plot(yeast_spectral_feature[i, :])
plt.xlabel('N Samples', fontsize=30)
plt.ylabel('Intensity', fontsize=30)
plt.title("Yeast Spectral Feature", fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(50,25))
plt.plot(ecoli_spectral_feature.mean(axis=0), linewidth = 8)
plt.plot(yeast_spectral_feature.mean(axis=0), linewidth = 8)
plt.legend(["E.Coli", "Yeast"], loc="best", fontsize = 30)
plt.xlabel('N Samples', fontsize=30)
plt.ylabel('Intensity', fontsize=30)
plt.title("Mean of Spectral Feature (E.Coli & Yeast)", fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(50,25))
plt.plot(ecoli_spectral_feature.std(axis=0), linewidth = 8)
plt.plot(yeast_spectral_feature.std(axis=0), linewidth = 8)
plt.legend(["E.Coli", "Yeast"], loc="best", fontsize = 30)
plt.xlabel('N Samples', fontsize=30)
plt.ylabel('Std', fontsize=30)
plt.title("Std of Spectral Feature (E.Coli & Yeast)", fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(50,25))
plt.plot(ecoli_spectral_feature.var(axis=0), linewidth = 8)
plt.plot(yeast_spectral_feature.var(axis=0), linewidth = 8)
plt.legend(["E.Coli", "Yeast"], loc="best", fontsize = 30)
plt.xlabel('N Samples', fontsize=30)
plt.ylabel('Variance', fontsize=30)
plt.title("Variance of Spectral Feature (E.Coli & Yeast)", fontsize=30)
plt.show()
```
